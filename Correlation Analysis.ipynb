```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\Lalitha\OneDrive\Desktop\project_spatial_datamining-main\data\cleaned\wq_lakes.csv' 
df = pd.read_csv(file_path)
```

```python
df = df.drop(columns=['StationNumber', 'SampleDatetime', 'RiverSubBasinCode'])
```

```python
df.isnull().sum()  # Check for missing values
```

```python
correlation_matrix = df.corr()  # Pearson correlation by default
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix on Water Quality")
plt.show()
```

Water Quality analysis from heatmap
1. pH vs Dissolved Oxygen = 0.68 -> Positive Correlation
2. pH vs Temperature = 0.59 -> Positive Correlation
3. Conductivity vs Dissolved Oxygen = -0.04 -> Negative Correlation
4. Dissolved oxygen is strongly linked to pH and temperature
5. Conductivity doesn’t significantly impact oxygen levels in water.
